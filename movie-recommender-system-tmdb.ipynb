```python
# Import the nltk library
import nltk

# Import the PorterStemmer
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
movi['tags'] = movi['tags'].apply(stems)
```

```python
movi.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Assuming movi['tags'] contains your text data
text_data = list(movi['tags'])

# Select the first 5000 documents
first_5000_text_data = text_data[:5000]

# Create the CountVectorizer with the specified parameters
cv = CountVectorizer(max_features=5000, stop_words='english')

# Fit and transform the first 5000 text data
vector = cv.fit_transform(first_5000_text_data).toarray()

# Print the shape of the resulting array
print("Shape of vector with first 5000 documents:", vector.shape)
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi = cosine_similarity(vector)
```

```python
simi.shape
```

```python
def recommend(movie):
    # Find the index of the given movie in the DataFrame
    index = movi[movi['title'] == movie].index[0]
    
    # Calculate similarity distances
    distances = sorted(list(enumerate(simi[index])), reverse=True, key=lambda x: x[1])
    
    # Print the recommended movies
    print("\033[1mRecommended Movies For You:\033[0m")  # Print in bold (ANSI escape sequence)
    for i in distances[1:20]:  # Skip the first one because it's the movie itself
        print(movi.iloc[i[0]].title)
```

```python
recommend('Batman')
```
